Q.10) Find the list of Affordable Restaurants.
The Criteria for Affordable Restaurants would be:
    1)Low Price  2)High Rated
    
First step will be to find the restaurants with average cost 1/4th the average cost of most expensive restaurant in our dataframe.

Let me explain:-The most expensive restaurant has an average meal cost= 6000. We'll try to stay economical and only pick the
restaurants that are 1/4th of 6000

```python
max_price = zomato_df['PRICE'].max()
one_fourth_price = max_price/4
one_fourth_price
```

```python
aff_rest_df = zomato_df[['NAME','PRICE','CUSINE_CATEGORY','REGION','CUSINE TYPE','RATING','VOTES']]
aff_rest_df = aff_rest_df[aff_rest_df['PRICE'] <= one_fourth_price]
aff_rest_df.sort_values(by = 'PRICE',inplace = True)
aff_rest_df
```

```python
highrate_rest_df=zomato_df[['NAME','PRICE','CUSINE_CATEGORY','REGION','CUSINE TYPE','RATING','VOTES']]
highrate_rest_df = highrate_rest_df[highrate_rest_df['RATING'] >= 4.5]
highrate_rest_df.sort_values(by = 'PRICE',inplace = True)
highrate_rest_df
```

```python
highrate_aff_df = pd.merge(aff_rest_df, highrate_rest_df, how = 'inner', on = ['NAME','REGION'])
highrate_aff_df = highrate_aff_df[['NAME','PRICE_x','CUSINE_CATEGORY_x','REGION','CUSINE TYPE_x','RATING_x','VOTES_x']]
highrate_aff_df.rename(columns = {'PRICE_x' : 'PRICE', 'CUSINE_CATEGORY_x' : 'CUSINE_CATEGORY',
                                  'CUSINE TYPE_x' : 'CUSINE TYPE','RATING_x' : 'RATING','VOTES_x':'VOTES'}, inplace = True)
highrate_aff_df
```

Q.11) Find the list of most Reliable Restaurants:
     
     The Criteria for most Reliable Restaurants would be:
         
     1)Low Price 2)High Rated 3)Large no of Votes

```python
mean_votes = zomato_df['VOTES'].mean()
mean_votes
```

```python
mean_rest_df = zomato_df[['NAME','PRICE','CUSINE_CATEGORY','REGION','CUSINE TYPE','RATING','VOTES']]
mean_rest_df = mean_rest_df[mean_rest_df['VOTES'] > mean_votes]
mean_rest_df.sort_values(by = 'VOTES',inplace = True)
mean_rest_df
```

```python
reliable_rest_df = pd.merge(highrate_aff_df,mean_rest_df, how = 'inner', on = ['NAME','REGION'])
reliable_rest_df = reliable_rest_df[['NAME','PRICE_x','REGION','CUSINE_CATEGORY_x','CUSINE TYPE_x','RATING_x','VOTES_x']]
reliable_rest_df.rename(columns = {'PRICE_x' : 'PRICE','CUSINE_CATEGORY_x' : 'CUSINE_CATEGORY','CUSINE TYPE_x' : 'CUSINE TYPE',
                                'RATING_x' : 'RATING', 'VOTES_x': 'VOTES','PRICE_x' : 'PRICE'}, inplace = True)
reliable_rest_df
```
