```python
import pandas as pd 
#pandas is library and it contains functions and classes.
#we import pandas here and take pandas as pd.
```

```python
dataset=pd.read_csv("Salary_Data.csv")
#dataset is a variable and assign salary data csv file to dataset variable.
```

```python
dataset
#run dataset
```

```python
independent=dataset[["YearsExperience"]]
independent
#independent is a variable and assign Years Experience which is also an input to independent
```

```python
dependent=dataset[["Salary"]]
dependent
#dependent is a variable and assign salary which is also an output to dependent
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(independent, dependent, test_size=0.30, random_state=0)
#sklearn is a library. take one class called model selection and function called train test split.
#take x_train, x_test, y_train, y_test and assign respective variables independent, dependent, train and test
#take train and test ratio as 7:3
```

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(x_train,y_train)
#take linear model class and import LinearRegression function from sklearn library.
#take variable regressor and assigned LinearRegression to regressor 
#Learning our model from x_train and x_test.
```

```python
weight=regressor.coef_
weight
#variable regressor contains all functions of trained model
#we can find value of weight which is slope of the linear equation.
```

```python
bias=regressor.intercept_
bias
#Also we can find the value of bias which is intercept of the linear equation.
```

```python
y_pred=regressor.predict(x_test)
#take variable as y_pred and assign the value of prediction of x_test.
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
#r2_score is a validating parameter and it is import from sklearn library.
#In r2_score validating parameter cross check the value of y_test and y_pred.
#y_pred is nothing but predicted x_test.
```

```python
r_score
#find the value r2_score.
#if value of r2_score model is nearly to 1. then our model is good.
```

```python
import pickle
#pickle is a library to save our model.
```

```python
filename="finalized_model.sav"
#take a variable called filename and assigned finalized model.sav to the variable called filename.
```

```python
pickle.dump(regressor,open(filename,'wb'))
#pickle.dump is a function and it is used for save our model in sav file format
```

```python
loaded_model=pickle.load(open(filename,'rb'))
#take variable called loaded model assign pickle.load functions to variable loaded model.
```

```python
result=loaded_model.predict([[15]])
#here we get prediction of loaded model 
```
