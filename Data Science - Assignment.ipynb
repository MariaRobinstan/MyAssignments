4)What kind of relation between mba_p and salary?

```python
Quan = ['sl_no','ssc_p','hsc_p','degree_p','mba_p','etest_p','salary']
Quan
```

```python
data = dataset[Quan]
data
```

```python
def calc_vif(data):
    from statsmodels.stats.outliers_influence import variance_inflation_factor
    vif = pd.DataFrame()
    vif['Variables'] = data.columns
    vif['VIF'] = [variance_inflation_factor(data.values,i) for i in range(data.shape[1])]
    return vif
calc_vif(data[['mba_p','salary']])
```

```python
#Here we get variance inflation values between mba_p and salary 2.707661.
#2.707661 < 5
#Therefore relation between mba_p and salary is moderately correlated.
```

```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols

model = ols('etest_p ~ mba_p + etest_p : mba_p', data = dataset).fit()

anova_table = sm.stats.anova_lm(model)
print(anova_table)
```

Null Hypothesis  = H0 : There is no significant difference between mba_p and etest_p
                   H0 : Mean1 = Mean2
            
Alternative Hypothesis = H1 : There is a significant difference between mba_p and etest_p

```python
#we compare the p-value with significance level to decide on our hypothesis:
#Since the p-value =  4.446395e-88 is less than 0.005, we reject the null hypothesis in favour of the alternative hypothesis.
```

```python
from scipy.stats import ttest_ind
SciTech = dataset[dataset['degree_t'] == 'Sci&Tech']['salary']
MktHR = dataset[dataset['specialisation'] == 'Mkt&HR']['salary']
ttest = ttest_ind(SciTech,MktHR)
print(ttest)
```

9)Converting the normal distribution to standard normal distribution for the salary column.

```python
import seaborn as sns

def StdNormalDistribution(column,dataset):
    mean = dataset[column].mean()
    std = dataset[column].std()
    values = [i for i in dataset[column]]

    z_score =[((x - mean)//std ) for x in values]

    sns.distplot(z_score, kde = True)
    z_mean = sum(z_score)/len(z_score)
    return z_mean

StdNormalDistribution('salary',dataset)
```

10)What is the probability density function of the salary range of 7,00,000 to 9,00,000

```python
import seaborn as sns

#sns.distplot(column,dataset,kde = True)

def findPDF(column,dataset,startrange,endrange):
    from matplotlib import pyplot
    from scipy.stats import norm
    
    ax = sns.distplot(dataset[column], kde = True, kde_kws = {'color' : 'blue'},color = 'Green')
    pyplot.axvline(startrange)
    pyplot.axvline(endrange,color = 'red')
    
    sample = dataset[column]
    sample_mean = sample.mean()
    sample_std = sample.std()
    print('Mean = %3f, Standard Deviation = %3f'%(sample_mean,sample_std))
    dist = norm(sample_mean, sample_std)
    
    values = [value for value in range(startrange,endrange)]
    probabilities = [dist.pdf(value) for value in values]
    prob = sum(probabilities)
    print('The area between range({},{}) : {}'.format(startrange, endrange, sum(probabilities)))
                                                    
    return prob*100

findPDF('salary',dataset,700000,900000)
    
```

```python
#The probability density function of the salary of range 700000 to 900000 is 0.059%
```

```python
import numpy as np

def calculate_Outliers(data, category_column, category_value, numerical_column):
    group = data[data[category_column] == category_value][numerical_column]
    
    Q0 = np.percentile(group, 0)
    Q1 = np.percentile(group, 25)
    Q2 = np.percentile(group, 50)
    Q3 = np.percentile(group, 75)
    Q4 = np.percentile(group, 100)
    Minimum = group.min()
    Maximum = group.max()
    IQR = Q3 - Q1
    Lesser_Outlier = Q1 - 1.5 * IQR
    Greater_Outlier = Q3 + 1.5 * IQR
    
    print(f'{category_value} Outliers for {numerical_column}:')
    print('(Minimum, Maximum): ', (Minimum, Maximum))
    print('Q0:',Q0, 'Q1:', Q1, ', Q2:', Q2, ', Q3:', Q3, ', Q4:', Q4)
    print('IQR: ', IQR)
    print('(Lesser Outlier, Greater Outlier): ', (Lesser_Outlier, Greater_Outlier))

calculate_Outliers(dataset, 'hsc_s', 'Commerce', 'hsc_p')
calculate_Outliers(dataset, 'hsc_s', 'Science', 'hsc_p')
calculate_Outliers(dataset, 'hsc_s', 'Arts', 'hsc_p')
```
